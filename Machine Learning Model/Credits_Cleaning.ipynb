```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import ast
import json
```

Se usará una forma similar a las usadas en Movies_ETL.ipynb, utilizando la librería `ast` para desanidar las columnas y poder separar los datos de actores y miembros del crew en dos datasets, `actors.csv` y `crew.csv`. Estos serviran para las funciones de la API donde debe retornarse a los actores junto con datos de las peliculas donde participaron, asi como el retorno de los directores y las peliculas que dirigieron.

```python
df = pd.read_csv("credits.csv") # el archivo credits.csv lo elimine luego de las transformaciones debido a su tamaño
df
```

```python
dataframe_from_col1 = pd.DataFrame(columns=['cast_id','character','credit_id','actor_gender','actor_id','actor_name','order','profile_path'])

for value in df["cast"]:
    if pd.notna(value):  # Check if the value is not NaN
        data = ast.literal_eval(value)
        if isinstance(data, list) and len(data) > 0:
            cast_id = [item['cast_id'] for item in data]
            character = [item['character'] for item in data]
            credit_id = [item['credit_id'] for item in data]
            actor_gender = [item['gender'] for item in data]
            actor_id = [item['id'] for item in data]
            actor_name = [item['name'] for item in data]
            order = [item['order'] for item in data]
            profile_path = [item['profile_path'] for item in data]
            dataframe_from_col1.loc[len(dataframe_from_col1)] = [
                cast_id,
                character,
                credit_id,
                actor_gender,
                actor_id,
                actor_name,
                order,
                profile_path
            ]
        else:
            dataframe_from_col1.loc[len(dataframe_from_col1)] = [None, None,None, None,None, None,None,None]
    else:
        dataframe_from_col1.loc[len(dataframe_from_col1)] = [None, None,None, None,None, None,None,None]

dataframe_from_col1.head(5)
```

```python
result_dataframe = df.join(dataframe_from_col1)
result_dataframe.head(5)
```

```python
dataframe_from_col2 = pd.DataFrame(columns=['crew_credit_id','crew_department','crew_gender','crew_id','crew_job','crew_name','crew_profile_path'])

for value in df["crew"]:
    if pd.notna(value):  # Check if the value is not NaN
        data = ast.literal_eval(value)
        if isinstance(data, list) and len(data) > 0:
            crew_credit_id = [item['credit_id'] for item in data]
            crew_department = [item['department'] for item in data]
            crew_gender = [item['gender'] for item in data]
            crew_id = [item['id'] for item in data]
            crew_job = [item['job'] for item in data]
            crew_name = [item['name'] for item in data]
            crew_profile_path = [item['profile_path'] for item in data]
            dataframe_from_col2.loc[len(dataframe_from_col2)] = [
                crew_credit_id,
                crew_department,
                crew_gender,
                crew_id,
                crew_job,
                crew_name,
                crew_profile_path
            ]
        else:
            dataframe_from_col2.loc[len(dataframe_from_col2)] = [None, None,None, None,None, None,None]
    else:
        dataframe_from_col2.loc[len(dataframe_from_col2)] = [None, None,None, None,None, None,None]

dataframe_from_col2.head(5)
```

```python
result_dataframe = result_dataframe.join(dataframe_from_col2)
result_dataframe.head(5)
```

```python
result_dataframe.drop(columns=["cast","crew"],axis=1,inplace=True)
result_dataframe.head(5)
```

```python
compressed_df = pd.read_csv("credits_unnested.csv",compression='gzip')
compressed_df.head(2)
```

Para poder desanidar cada columna se usará la función `explode()` de pandas. Pero se debe primero aplicar una `lambda` a cada columna que se quiere desanidar

```python
df = compressed_df.copy()
for item in df.columns:
  if item != 'id':
    df[item] = df[item].apply(lambda x: ast.literal_eval(x) if not pd.isna(x) else [])
df.head()
```

Ahora que se convirtieron a listas usando `ast.list_eval`, se debe hacer un `explode` de cada columna como sigue

```python
df.columns
['id', 'cast_id', 'character', 'credit_id', 'actor_gender', 'actor_id',
       'actor_name', 'order', 'profile_path', 'crew_credit_id',
       'crew_department', 'crew_gender', 'crew_id', 'crew_job', 'crew_name',
       'crew_profile_path'],
```

```python
cast_id = df[["id","cast_id"]].explode("cast_id").reset_index(drop=True)
character= df[["id","character"]].explode("character").reset_index(drop=True)
credit_id= df[["id","credit_id"]].explode("credit_id").reset_index(drop=True)
actor_gender= df[["id","actor_gender"]].explode("actor_gender").reset_index(drop=True)
actor_id= df[["id","actor_id"]].explode("actor_id").reset_index(drop=True)
actor_name= df[["id","actor_name"]].explode("actor_name").reset_index(drop=True)
order= df[["id","order"]].explode("order").reset_index(drop=True)
profile_path= df[["id","profile_path"]].explode("profile_path").reset_index(drop=True)
crew_credit_id= df[["id","crew_credit_id"]].explode("crew_credit_id").reset_index(drop=True)
crew_department= df[["id","crew_department"]].explode("crew_department").reset_index(drop=True)
crew_gender= df[["id","crew_gender"]].explode("crew_gender").reset_index(drop=True)
crew_id= df[["id","crew_id"]].explode("crew_id").reset_index(drop=True)
crew_job= df[["id","crew_job"]].explode("crew_job").reset_index(drop=True)
crew_name= df[["id","crew_name"]].explode("crew_name").reset_index(drop=True)
crew_profile_path= df[["id","crew_profile_path"]].explode("crew_profile_path").reset_index(drop=True)
```

```python
new_df = cast_id
new_df['character'] = character["character"]
new_df['credit_id'] = credit_id["credit_id"]
new_df['actor_gender'] = actor_gender["actor_gender"]
new_df['actor_id'] = actor_id["actor_id"]
new_df['actor_name'] = actor_name["actor_name"]
new_df['order'] = order["order"]
new_df['profile_path'] = profile_path["profile_path"]
new_df['crew_credit_id'] = crew_credit_id["crew_credit_id"]
new_df['crew_department'] = crew_department["crew_department"]
new_df['crew_gender'] = crew_gender["crew_gender"]
new_df['crew_id'] = crew_id["crew_id"]
new_df['crew_job'] = crew_job["crew_job"]
new_df['crew_name'] = crew_name["crew_name"]
new_df['crew_profile_path'] = crew_profile_path["crew_profile_path"]
new_df.head()
```

```python
final_credits_dataset = new_df.copy()
final_credits_dataset.to_csv("credits_dataset_final.csv",index=False,compression="gzip")
```

```python
# primero se convierten los datos con ast.list_eval
df = df.copy()
for item in df.columns:
  if item != 'id':
    df[item] = df[item].apply(lambda x: ast.literal_eval(x) if not pd.isna(x) else [])
df.head()
```

```python
cast_id = df[["id","cast_id"]].explode("cast_id").reset_index(drop=True)
character= df[["id","character"]].explode("character").reset_index(drop=True)
credit_id= df[["id","credit_id"]].explode("credit_id").reset_index(drop=True)
actor_gender= df[["id","actor_gender"]].explode("actor_gender").reset_index(drop=True)
actor_id= df[["id","actor_id"]].explode("actor_id").reset_index(drop=True)
actor_name= df[["id","actor_name"]].explode("actor_name").reset_index(drop=True)
order= df[["id","order"]].explode("order").reset_index(drop=True)
profile_path= df[["id","profile_path"]].explode("profile_path").reset_index(drop=True)
crew_credit_id= df[["id","crew_credit_id"]].explode("crew_credit_id").reset_index(drop=True)
crew_department= df[["id","crew_department"]].explode("crew_department").reset_index(drop=True)
crew_gender= df[["id","crew_gender"]].explode("crew_gender").reset_index(drop=True)
crew_id= df[["id","crew_id"]].explode("crew_id").reset_index(drop=True)
crew_job= df[["id","crew_job"]].explode("crew_job").reset_index(drop=True)
crew_name= df[["id","crew_name"]].explode("crew_name").reset_index(drop=True)
crew_profile_path= df[["id","crew_profile_path"]].explode("crew_profile_path").reset_index(drop=True)
```

```python
actors_df = cast_id
actors_df['character'] = character["character"]
actors_df['credit_id'] = credit_id["credit_id"]
actors_df['actor_gender'] = actor_gender["actor_gender"]
actors_df['actor_id'] = actor_id["actor_id"]
actors_df['actor_name'] = actor_name["actor_name"]
actors_df['order'] = order["order"]
actors_df['profile_path'] = profile_path["profile_path"]

crew_df = crew_credit_id
crew_df['crew_department'] = crew_department["crew_department"]
crew_df['crew_gender'] = crew_gender["crew_gender"]
crew_df['crew_id'] = crew_id["crew_id"]
crew_df['crew_job'] = crew_job["crew_job"]
crew_df['crew_name'] = crew_name["crew_name"]
crew_df['crew_profile_path'] = crew_profile_path["crew_profile_path"]
crew_df.head()
```
